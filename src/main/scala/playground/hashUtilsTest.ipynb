```python
import numpy as np
import math
from collections import Counter
```

```python
def hash_func(pc: int, nEntries: int) -> int:
    indexBits = int(math.ceil(math.log2(nEntries)))
    part1 = (pc >> 4) & 0xF
    part2 = (pc >> 8) & 0xF
    part3 = (pc >> 16) & 0xF
    part4 = (pc >> 20) & 0xF
    mixed = part1 ^ part2 ^ part3 ^ part4
    return mixed & ((1 << indexBits) - 1)

nEntries = 8  # 你可以修改为其他值
results = []
num_tests = 1000

for i in range(num_tests):
    pc = i * 8
    idx = hash_func(pc, nEntries)
    results.append(idx)
    print(f"{pc} ----> {idx}")

print(f"不同结果数量: {len(results)} / {nEntries}")
Counter_results = Counter(results)
Counter_results
```
